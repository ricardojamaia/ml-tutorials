# The Ultimate Hello World of Neural Networks in Tensorflow 2

When we learn a new programming language we normally start by some basic "Hello World!" example. This tutorial guides you through a basic Neural Network (NN) example in Tensorflow. Like any "Hello World!" example, it doesn't tell you the whole story but allow you to start your learning path.



# The Problem

A cab company charges a 2.5 EUR boarding rate in addition to its meter which is 2 EUR for every km.

In this example we will build a NN to predict how much a taxi ride costs given the number of km.

Obviously, this is a simple problem which could be easily translated into a linear equation: y = m * x + b
But, instead of providing the rules to get data as we do in tradicional programming, we will provide data so that the NN learns the rules. This is how macinhe learning works. You provide data and computer learns the rules from that *data*. 

# The solution

In order to predict the cost of a taxi ride will will build a very basic Neural Network using Tensorflow 2.

We will start by importing the required python modules and then:
- Create the Neural Network
- Config the model with optimizer and loss with **`model.compile()`**
- Train the model with **`model.fit()`**
- Use the model to do prediction with **`model.predict()`**.


## Imports

The following packages will be required:
- **Tensorflow**: an open source machine learning framework developed by Google Brain.
- **Keras**: a deep learning high-level API written in Python, running on top of TensorFlow
- **Numpy**: a porwerfull and well known general-purpose array-processing and scientic computing package.

```
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

## Creating the Neural Network

Next we will create the most basic Neural Network possible. It will be composed by one layer with one Neuron which receives one input.

```
model = tf.keras.Sequential([
        keras.layers.Dense(1)
    ])
```

# Configure the model

Several iterations will be done when training the Neural Network. In each iteration the network computes predictions (in our case the predicted cost of the taxi ride) for each trainning sample given the input values (number of km).

Then, it computes how far these predicted values are from the expected values.  This is given by a so called loss function.

Finally, it updates the internal parameters (***weights*** and ***bias***) in order to minimize the loss. The optimizer will determine how these parameters are updated.

When configuring the model we need to specify a loss function and the optimization algorithm.

The following picture illustrates our Neural Network with only one node. Don't worry if at this stage it doesn't make much sense to you.

neuron-one-input.png

```
model.compile(optimizer='sgd', loss='mean_squared_error')
```

## Providing Training Data

Now we need to provide our NN training data. We will build an array with the inputs and another one with the corresponding expected outputs. We will use Numpy for that.

From our problem definition we know that `y = 2 * x + 2.5`

```
x_train = np.array([4.2, 5.5, 10.0, 12.3, 8.7, 9.2, 14.5], dtype=float)
y_train = np.array([10.9, 13.5, 22.5, 27.1, 19.9, 20.9, 31.5], dtype=float)
```

## Training the model

Now it's time to train our Neural Network. Here is where the *learning magic* occurs. As describe above several iterations will be performed in which weights (only one in our case) and bias will be adjusted in an attempt to bring predictions closer to expected values.

Each iteration over the full training set is called an ***epoch***.

```
model.fit(x_train, y_train, epochs=500)
```

## Predict

Now it's time to test our Neural Network. Let's predict how much we would pay for 10km cab ride.

```
print(model.predict([11.0]))
```
